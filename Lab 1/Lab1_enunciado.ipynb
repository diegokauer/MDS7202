<code>Escriba sus respuestas acá:</code>

11. `git stash` sirve para limpiar y archivar un directorio y volver a su versión más reciente (El commit más reciente en el `HEAD`). Se puede restaurar la versión archivada con `git stash apply`.

12. `git reset` cambia el `HEAD` a un commit especificado perdiendo la historia entre el último commit y el commit al que se movió el `HEAD`. Por otra parte, lo que hace el `git revert` es volver al estado del commit seleccionado revertiendo todos los cambios, pero sin modificar el `HEAD`, por lo que se mantiene la historia de commits y se agrega uno nuevo con los cambios deshechos.
