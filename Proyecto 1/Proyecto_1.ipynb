```python
import os
import datetime
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer, MinMaxScaler, LabelEncoder, OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.ensemble import IsolationForest
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import warnings
import shutil
warnings.filterwarnings("ignore")

class Profiler():
    def __init__(self, df, scale_method=None, cluster_method=None, anomalies_method=None):
        self.data = df.copy()
        self.scale_method = scale_method
        self.cluster_method = cluster_method
        self.anomalies_method = anomalies_method
        today = datetime.date.today().strftime("%d-%m-%Y")
        dir_name = f'EDA_{today}'
        self.dir_name = dir_name
        if not os.path.exists(dir_name):
            os.makedirs(dir_name)
        self.clean_transformer = None
        self.scaler_transformer = None
    
    def summarize(self, columns=None, custom_threshold=None):
        if columns is None:
            columns = self.data.columns

        summary_results = []
        
        #Contar los duplicados
        duplicados = self.data.duplicated().sum()
        print(f'N√∫mero de Datos Duplicados: {duplicados}')
        print(f'% de Datos Duplicados: {(duplicados / len(self.data) * 100):.2f}%')
        print('\n')
        
        for column in columns:
            # Se revisa si las columnas filtradas pertenecen al dataset
            if column not in self.data.columns:
                print(f"Warning: {column} no encontrada en el dataset.")
                continue
            column_data = self.data[column]
            variable_type = str(column_data.dtypes)
            unique_count = column_data.nunique()
            missing_count = column_data.isnull().sum()

            if variable_type == 'object':
                print(f'Variable: {column} (Categorica)')
                print(f'N√∫mero de Valores √önicos: {unique_count}')
                print(f'% de Valores Faltantes: {(missing_count / len(column_data) * 100):.2f}%')
                summary_results.append(f'Variable: {column} (Categorica)')
                summary_results.append(f'N√∫mero de Valores √önicos: {unique_count}')
                summary_results.append(f'% de Valores Faltantes: {(missing_count / len(column_data) * 100):.2f}%')
            else:
                print(f'Variable: {column} (Categorica)')
                print(f'N√∫mero de Valores √önicos: {unique_count}')
                print(f'% de Valores Faltantes: {(missing_count / len(column_data) * 100):.2f}%')
                print(f'Valor M√≠nimo: {column_data.min()}')
                print(f'Valor M√°ximo: {column_data.max()}')
                print(f'Promedio: {column_data.mean()}')
                print(f'Percentil 25: {column_data.quantile(0.25)}')
                print(f'Percentil 50 (Mediana): {column_data.median()}')
                print(f'Percentil 75: {column_data.quantile(0.75)}')
                summary_results.append(f'Variable: {column} (Num√©rica)')
                summary_results.append(f'N√∫mero de Valores √önicos: {unique_count}')
                summary_results.append(f'% de Valores Faltantes: {(missing_count / len(column_data) * 100):.2f}%')
                summary_results.append(f'Valor M√≠nimo: {column_data.min()}')
                summary_results.append(f'Valor M√°ximo: {column_data.max()}')
                summary_results.append(f'Promedio: {column_data.mean()}')
                summary_results.append(f'Percentil 25: {column_data.quantile(0.25)}')
                summary_results.append(f'Percentil 50 (Mediana): {column_data.median()}')
                summary_results.append(f'Percentil 75: {column_data.quantile(0.75)}')

                # Detecci√≥n de anomal√≠as en base a un umbral determinado (ajustable)
                if custom_threshold is not None:
                    outliers = sum((column_data - column_data.mean()).abs() > custom_threshold * column_data.std())
                    print(f'N¬∞ de outliers: {outliers}')
                    summary_results.append(f'N¬∞ de outliers: {outliers}')
            print('\n')
            summary_results.append('\n')

        summary_file = os.path.join(self.dir_name, 'summary.txt')

        with open(summary_file, 'w') as file:
            file.write('\n'.join(summary_results))

        print(f'El resumen est√° guardado en: {summary_file}')

    def plot_vars(self, columns=None, top_categories=None):
        # Creaci√≥n de la carpeta EDA_fecha/plots
        plots_dir = os.path.join(self.dir_name, 'plots')
        if 'plots' not in os.listdir(self.dir_name):
            os.makedirs(plots_dir)
        
        if columns is None:
            columns = self.data.columns

        for column in columns:
            # Se revisa si las columnas filtradas pertenecen al dataset
            if column not in self.data.columns:
                print(f"Warning: {column} no encontrada en el dataset.")
                continue

            column_data = self.data[column]
            variable_type = str(column_data.dtypes)

            if variable_type == 'object':
                if top_categories is not None:
                    top_categories_count = column_data.value_counts().head(top_categories)
                    plt.figure(figsize=(10, 6))
                    sns.barplot(x=top_categories_count.index, y=top_categories_count.values)
                    plt.title(f'Top {top_categories} categor√≠as de variable {column}')
                    plt.xticks(rotation=45)
                    plt.show()
                    plt.tight_layout()

                    # Guardar el gr√°fico en formato PDF
                    plot_file = os.path.join(plots_dir, f'{column}.pdf')
                    plt.savefig(plot_file)
                    plt.show()
                    plt.close()

            else:
                # Gr√°fico de distribuci√≥n de densidad
                plt.figure(figsize=(10, 6))
                sns.kdeplot(column_data, shade=True)
                plt.title(f'Gr√°fico de distribuci√≥n de densidad de {column}')
                plt.tight_layout()
                plt.show()

                # Guardar el gr√°fico en formato PDF
                plot_file = os.path.join(plots_dir, f'{column}.pdf')
                plt.savefig(plot_file)
                plt.close()
            
        # Filtrar las columnas num√©ricas
        numeric_columns = self.data.select_dtypes(include=['number']).columns
        
        # Generar la matriz de correlaci√≥n para las columnas num√©ricas
        correlation_matrix = self.data[numeric_columns].corr()
        plt.figure(figsize=(12, 8))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
        plt.title('Matriz de correlaci√≥n')
        plt.tight_layout()
        plt.show()

        # Guardar la matriz de correlaci√≥n en formato PDF
        corr_plot_file = os.path.join(plots_dir, 'matriz_correlacion.pdf')
        plt.savefig(corr_plot_file)
        plt.close()
        
        ### Graficar coeficiente V de Cramer
        ###
        ###
        
        print(f'Los gr√°ficos est√°n guardados en: {plots_dir}')

    def clean_data(self, drop_duplicates=True, deal_na=None):
        # Creaci√≥n de la carpeta EDA_fecha/plots
        clean_dir = os.path.join(self.dir_name, 'clean_data')
        if 'clean_data' not in os.listdir(self.dir_name):
            os.makedirs(clean_dir)
        
        # Definir la funci√≥n de limpieza
        def cleaning_function(data, drop_duplicates, deal_na):

            if drop_duplicates:
                data = data.drop_duplicates()
            
            # Separar columnas no at√≥micas
            columna_no_atomica = []
            for columna in data.columns:
                if '-' in columna:
                    columna_no_atomica.append(columna)
                    
            # Define el car√°cter especial con el que deseas reemplazar los caracteres especiales
            caracter_reemplazo = '-'
                    
            # Separar el nombre de la columna no at√≥mica en las columnas at√≥micas correspondientes       
            columnas_nuevas = re.split(r'[-_]', columna_no_atomica[0])
            
            # Funci√≥n para reemplazar caracteres especiales en una cadena
            def reemplazar_caracteres_especiales(cadena):
                # Utiliza una expresi√≥n regular para encontrar todos los caracteres especiales excepto los puntos y reemplazarlos
                return re.sub(r'[^A-Za-z0-9.]+', caracter_reemplazo, cadena)
            
            # Aplica la funci√≥n a la columna no at√≥mica del DataFrame      
            data[columna_no_atomica[0]] = data[columna_no_atomica[0]].apply(reemplazar_caracteres_especiales)
            data[columnas_nuevas] = data[columna_no_atomica[0]].str.split(caracter_reemplazo, expand=True)
                    
            # Funci√≥n personalizada para tratar 'nan' como NaN al convertir a tipo num√©rico
            def to_numeric_and_nan(valor):
                if valor.lower() == 'nan':
                    return np.nan
                try:
                    return pd.to_numeric(valor)
                except:
                    return valor  # Mant√©n el valor si no se puede convertir
                    
            for columna in columnas_nuevas:
                data[columna] = data[columna].apply(to_numeric_and_nan)
                    
            data = data.drop(columna_no_atomica[0], axis=1)
            
            if deal_na == "drop":
                data = data.dropna()
            elif deal_na == "impute":
                for column in data.columns:
                # Revisa si se trata de una columna num√©rica, para hacer un impute de medias
                    if pd.api.types.is_numeric_dtype(data[column]):
                        # Imputar valores nulos en columnas num√©ricas con la media
                        data[column].fillna(data[column].mean(), inplace=True)
                    else:
                        # Imputar valores nulos en columnas categ√≥ricas con "Unknown"
                        data[column].fillna("Unknown", inplace=True)
            
            return data
        
        # Crear un FunctionTransformer con la funci√≥n de limpieza
        cleaner = FunctionTransformer(cleaning_function, kw_args={'drop_duplicates': drop_duplicates, 'deal_na': deal_na})

        self.clean_transformer = cleaner
        
        # Aplicar la transformaci√≥n a los datos
        data = cleaner.fit_transform(self.data)
        
        # Guardar los datos procesados en un archivo CSV
        data_limpia_dir = os.path.join(clean_dir, 'data.csv')
        data.to_csv(data_limpia_dir, index=False)
        
        print(f'La data limpia est√° en: {data_limpia_dir}')
        
        return data

        
    def scale(self, method):
        # Crear directorio
        dir_name = os.path.join(self.dir_name, 'scale')
        if not os.path.exists(dir_name):
            os.makedirs(dir_name)
            
        self.clean_data(drop_duplicates=True, deal_na='impute')
        scale_df = self.clean_transformer.fit_transform(self.data)

        # Obtener columnas numericas y no-numericas
        num_scale_df = scale_df.select_dtypes(exclude='object')
        obj_scale_df = scale_df.select_dtypes(include='object')

        class CategoricalEncoder(BaseEstimator, TransformerMixin):
            def __init__(self, method):
                self.method = method
                self.encoders = None
                
            def fit(self, X, y=None):
                if isinstance(method, LabelEncoder):
                    encoders = [LabelEncoder().fit(X.loc[:,col]) for col in X.columns]
                    self.encoders = encoders
                    self.columns = X.columns
                else:
                    self.method.fit(X)
                return self

            def transform(self, X, y=None):
                if self.encoders:
                    X = np.stack(tuple(e.transform(X[col]) for e, col in zip(self.encoders, X.columns)), axis=1)
                else:
                    X = self.method.transform(X)
                return X

            def get_feature_names_out(self, a):
                if self.encoders:
                    return self.columns
                return self.method.get_feature_names_out()

        
        # Transformaciones
        def log_minmax(x):
            scaler = MinMaxScaler()
            return scaler.fit_transform(np.log(x + 1))

        scaler = ColumnTransformer(
            [
                ('method', CategoricalEncoder(method), obj_scale_df.columns),
                ('log_minmax', FunctionTransformer(log_minmax, feature_names_out='one-to-one'), num_scale_df.columns),
                # ('minmax', MinMaxScaler(), num_scale_df.columns),
            ], remainder='passthrough'
        )

        # Ajustar datos
        scale_data = scaler.fit_transform(scale_df)
        self.scaler_transformer = scaler

        # Construccion df y guardado
        scale_df = pd.DataFrame(scale_data, columns=scaler.get_feature_names_out())
        scale_df.to_csv(os.path.join(dir_name, 'scaled_features.csv'))

        print(f'La data scaled est√° en: {os.path.join(dir_name, "scaled_features.csv")}')
        
        return scale_df


    def make_cluster(self, cluster_algo, raw_data=None):
        # Crear directorio
        dir_name = os.path.join(self.dir_name, 'clusters')
        if not os.path.exists(dir_name):
            os.makedirs(dir_name)

        # Obtener datos
        if not raw_data is None:
            self.data = raw_data
            self.scale(self.scale_method)
        else:
            if self.scaler_transformer is None:
                self.scale(self.scale_method)
        data = self.data
        
        # Busqueda de n mediante el metodo del cono
        class CodoKNN(BaseEstimator, TransformerMixin):
            def __init__(self, method):
                self.method = method
                
            def fit(self, X, y=None):
                # Metodo del codo
                inertias = [
                    [i, KMeans(n_clusters=i, random_state=0, n_init='auto').fit(X).inertia_]
                    for i in range(1, 12)]
                inertias = pd.DataFrame(inertias, columns=["n", "inertia"])
                inertias['pct_change'] = inertias['inertia'].pct_change()
                promedios = []
                for i in range(len(inertias)):
                    promedio_fila = inertias.iloc[i:,:]['pct_change'].mean()
                    promedios.append(promedio_fila)
                inertias['promedios'] = promedios
                n = int(inertias.iloc[np.argmax(inertias[inertias.promedios < -0.01].promedios)].n)
                self.method = self.method(n_clusters=n).fit(X)
                return self

            def transform(self, X, y=None):
                X = self.method.transform(X)
                self.labels_ = self.method.labels_
                return X
        

        # Pipe
        pipe = Pipeline([
            ('clean', self.clean_transformer),
            ('scale', self.scaler_transformer),
            ('cluster', CodoKNN(cluster_algo)),
            ('PCA', PCA(n_components=2))
        ], verbose=True)

        # Ajustar datos
        data = pipe.fit_transform(data)

        # Creaci√≥n de df
        data = pd.DataFrame(data, columns=['pca1', 'pca2'])
        data['cluster'] = pipe['cluster'].labels_

        # Grafico
        fig = sns.scatterplot(data, x='pca1', y='pca2', hue='cluster', palette='tab10').set_title(
            'Clusters de datos'
        )
        plt.show()
        fig = fig.get_figure()
        
        # Guardado csv y imagen
        fig.savefig(os.path.join(dir_name, 'clusters.png'))

        data_cluster = self.clean_data()
        data_cluster['cluster'] = data['cluster']
        data_cluster.to_csv(os.path.join(dir_name, 'data_clusters.csv'))

        print(f'La data de los clusters est√° en: {os.path.join(dir_name, "data_clusters.csv")}')
        print(f'El grafico est√° en: {os.path.join(dir_name, "clusters.png")}')
    
    def detect_anomalies(self, detect_anomalies_algo, raw_data=None):
        dir_name = os.path.join(self.dir_name, 'anomalies')
        if not os.path.exists(dir_name):
            os.makedirs(dir_name)

        # Obtener datos
        if not raw_data is None:
            self.data = raw_data
        self.scale(self.scale_method)
        data = self.data

        class DetectAnomalies(BaseEstimator, TransformerMixin):
            def __init__(self, method):
                self.method = method
                
            def fit(self, X, y=None):
                self.method.fit(X)
                return self

            def transform(self, X, y=None):
                self.labels_ = self.method.predict(X)
                return X

        pipe = Pipeline([
            ('clean', self.clean_transformer),
            ('scale', self.scaler_transformer),
            ('anomalies', DetectAnomalies(detect_anomalies_algo())),
            ('PCA', PCA(n_components=2))
        ], verbose=True)


        data = pipe.fit_transform(data)
        
        data = pd.DataFrame(data, columns=['pca1', 'pca2'])
        data['outlier'] = pipe['anomalies'].labels_

        data_an = self.clean_data()
        data_an['outlier'] = data['outlier']
        data_an.to_csv(os.path.join(os.path.join(dir_name, 'data_anomalies.csv')))
        
        fig = sns.scatterplot(data, x='pca1', y='pca2', hue='outlier', palette='tab10').set_title(
            'Detecci√≥n de anomalias en los datos'
        )
        plt.show()
        fig = fig.get_figure()
        fig.savefig(os.path.join(dir_name, 'anomalies.png'))

        print(f'La data de las anomalias est√° en: {os.path.join(dir_name, "data_anomalies.csv")}')
        print(f'El grafico est√° en: {os.path.join(dir_name, "anomalies.png")}')

    def profile(self):
        self.summarize()
        self.plot_vars()
        self.make_cluster(self.cluster_method)
        self.detect_anomalies(self.anomalies_method)

    def clearGarbage(self):
        shutil.rmtree(self.dir_name)
```
